## 1.1 审视你的资源
首先，我不希望你去买很高级的设备或者资源。他们大多在9000多元起步。实际上，若你作为爱好者或者感兴趣的参与者的话，你应该去审视你周围的资源。这个技巧非常有用，当你真正从事人工智能时，你需要评估你所需的资源。  
回答以下问题非常有用  


1.   你目前拥有的资源是什么？种类？有多少？
2.   你的项目需求是什么？要多少？有没有你无法提供的？
3.   你周围有什么资源？怎么获得？难度有多少？
4.   你力所能及的范围内有什么资源？质量如何？难度有多少？

回答上述问题后，填成一张表格，这将会在整本书都非常有用

## 1.2 你的硬件
当我们说硬件的时候，我们是指能提供算力的设备。下面列出了几种常见可用的硬件资源。
### 显卡
显卡是人们最容易获得的高算力设备。虽然它天生用于计算图形，但2013年的成功让人们开始看到显卡那巨大的潜能。
![alt text](https://raw.githubusercontent.com/ggPascal/PYCMS-STEP-INTO/master/285D7F3A-84FB-4C73-A42D-A1DD90C06537.png)  








业界普遍使用英伟达的 V100（2018年的显卡）作为主要算力，当然他过于昂贵，但是计算性能十分强大。V100 拥有张量处理核心和上万个CUDA处理核心，这两个处理核心我们之后会逐一介绍。  

个人领域的话英伟达RTX系列是最好的，也具有CUDA核心和张量处理核心。RTX 2070是2019年的推荐，往后的年份请自行查阅。  

![alt text](https://raw.githubusercontent.com/ggPascal/PYCMS-STEP-INTO/master/7B3042FB-6EA1-4C14-8A52-0E74515FC401.png)





但是2070还是有点贵，因此另一种选择是英伟达GTX 1060，没有张量处理核心，CUDA核心数量也有所减少。


---


AMD公司也提供很好的显卡，通常他们的价格更低。但是他们需要学习掌握OpenCL，因此不推荐初学者使用。因为初学者可能无法正确的将OpenCL编译进人工智能框架里。2019年推荐为Rayon系列的显卡，往后年份请自行查阅

### 边缘计算设备
这是一类小型的计算设备，一般用于提供更为廉价而又有针对性的算力。 

一般非专业的爱好者，选择Raspberry Pi的最新型号即可，他带有基于Linux的系统，缺点是寿命较短。大部分在书中提到的软件均能在上面运行。

专业的边缘计算设备可以去到英伟达，AMD，华为等企业官网查询。他们也有针对特定种类任务的计算设备。  

### 图形计算器  
是的，你没有看错！计算器可以作为数据预处理的平台，降低之后建立数据集和其它操作所需的算力。  
另一件较为疯狂的事情便是在计算器上运行神经网络。  

这方面推荐使用德州仪器或者CASCIO的图形计算器，记住检查是否有表格储存功能，与计算机是否有连接功能。若要进行更高级的操作，你还需要检查微积分功能的完整程度，是否含有CAS（计算机代数系统）。

我个人在2019年持有的是德州仪器 Ti-84 CE

### 中央处理器（CPU）
中央处理器也是一种常见的算力资源，但是，这只能起到辅助作用。  

英特尔在这方面做的很好，你可以放心的使用他的产品。i7是一个较好的选择，如果你能有幸使用到第十代芯片，你也可以尝试一下他的AI加速。  
AMD公司同样也可以提供很好的中央处理器，但是请注意：他们并未能很好的区分产品线，因此购买之前最好查询好相关信息。

### 安卓手机
安卓最大的好处便在于我们可以自定义我们的手机。这是苹果手机无法达到的。  
AidLearning是一个开源的深度学习工具包，你可以在他们的官网上下载安装包，最后再花3-5分钟，让它联网下载完相关数据包。  
里边提供的软件包能运行本教材的大部分代码，但是能否成功运行还是取决于手机的性能如何。  

软件是基于Termux实现的，它本身也是一个用于模拟Linux环境的APP，推荐也把它安装了，增强扩展性。  

## 1.2 软件
作为一个有常识的现代人，你应该清楚：复杂的运算不是人脑的优势项目。
拥有好的软件可以使你不容易走弯路。
## 后端计算框架
你的计算机在没有特定编程时无法计算向量，因此你需要一个后端来帮助你实现相关的流程  
下面的后端都是较为常用的框架。
### Tensorflow
谷歌开发的向量计算框架，拥有流式数据处理的特点。
兼容性很好，因此本教材将会采用Tensorflow作为基本后端使用

支持主流的编程语言。

### Theano
一款开源的向量计算框架，在很多平台可以良好的运行。  
本身也是历史较为久远的现代链式推导积分运算框架。  

### CMLK
微软开发的一套向量计算框架，主打计算机感知领域
因此它封装了一些用于计算机感知的加速方法

## 前段人性化封装库
如果采用后端直接实现相关的流程，你将会很耗费时间。
以下的两个前段是使用频率较高的两个前段。
### Keras
在近年兴起的人工智能前段封装库，主打积木式构建
近年来有较多的论文研究者会采用此框架，简化相关代码。

本书将采用基于Tensorflow为后端的Keras

### Pytroh
Facebook开发的人工智能框架，主打科学化开发  
该框架使用函数式语言，让开发者能以更好的方式设计一个科学的网络。

## 加速协议
如果不使用相关的加速协议，你的电脑也很难处理成千上万的数据。

### OpenBASCL
全称计算机基础线性代数库，这是一个高效的线性处理库，有效减少了数据错误所造成的校验处理时间

性能提升在5%左右

### cuDDN
一款专用于NVIDA的显卡原语优化库，他可以替代原有的显卡计算图像指令，使其更合适向量计算

性能提升取决于使用的NVIDA显卡性能如何，通常可提升80%

## 通讯协议
为了让其它设备也能计算相关数据，你需要了解以下两种协议

### CUDA
专用于NVIDA公司的计算通讯协议，范围包括光线追踪，向量计算，数据批处理等领域

注意只能用于NVIDA的设备，因此该协议一直是NVIDA公司的王牌

### OpenCL
可用于几乎任何一台可计算设备的协议，由多家公司联合开发。其中也包括了英伟达和AMD。因此这个协议可以大幅度降低算力成本，更好的利用周围的资源。也可以实现CPU与显卡一起计算，这是CUDA无法实现的

但是它很难进行编程，这是所有人应该要注意的。

## 编程语言
所有软件的基础，便是编程语言。以下列出了拥有深度学习编程能力的语言。  
### Python
一种现代化的固定格式语言，拥有很高的可读性。不过需要注意的是：本身语言并没有太过于强大的能力，需要安装Wheel（有时也被称之为“轮子”）实现特定的功能。  

由于是解释型语言，因此也会出现效率低下的问题。

本书的代码示例均采用Python作为代码。
### C++和C
两者都是非常有名的编程语言，都有非常可观的效率。  
但是在处理数组指针方面缺乏安全的方案。


都需要一定的基础，也支持.h为基础的扩展组件。能生成程序。
### Java
现代的翻译型语言，由一个称为“Java虚拟机”的环境去运行程序，可移植度很高。这也是面向安卓平台必须学会的编程语言。
### Tensorflow.JS
Tensorflow框架在JS语言上的移植，可以很好的处理边缘资源（浏览器）。  

这只是JS的移植，因此需要有JS语言基础

##其他算力资源和扩展工具
有的时候，你无法找到上述的资源，那你可能就要寻求互联网了。  
扩展工具可以加速你的开发。
### 云服务
云服务是一种较为低廉的算力资源，有免费和收费之分。  

免费的用于深度学习的产品估计也就只有谷歌的colab了，网页是用Jupyter Notebook呈现的，默认安装了Tensorflow，Keras等扩展。支持Ubuntu命令操作。可以免费使用TPU和GPU资源。每个笔记本大小上限是2MB

收费的云计算服务推荐亚马逊的GPU云。收费稳定在1～2美元/小时，使用之前最好先查询一下价格，做好计划。  
最好不要将收费云计算服务作为训练用平台，有些模型可能要几天才能搞定。

### Jupyter Notebook
一种代码笔记本，其优点是能以一种非常方便的形式和你的代码做互动。  
这也是为什么本教材使用该平台编写，因为这样能提供更好的互动性。
这是一款网页应用，也就是说手机也可以通过局域网访问的方式来参与编写。  
附录2 提供了如何安装和设置Jupyter Notebook

### Vscode
一款免费的代码编辑器，由微软公司开发。支持多种语言，也可以做文本编辑器使用。  
最大的特点是可以安装扩展组件，可以直接连接Github的特性也使它好评如潮。

### VS IDE
微软官方的IDE开发工具包，支持Java，C，C++，Python的开发。使用此平台可以很快的开发出为Windows平台而优化的程序。  

分为社区版（免费）和商业版（收费），安装完后会占用2GB左右（假设只安装一种语言支持）  
因此不建议使用IDE进行模型开发。

## 系统
似乎看上去，Windows就够了。但是，Linux可以更好的发挥电脑的性能。  
有些软件包只在某一个平台有用。

### Ubuntu
推荐使用的Linux发行版本，使用习惯类似于Windows和Mac系统。  
拥有多个硬件平台的编译版本

### Kali Linux
Debian发行系在针对网络安全方面的系统。  
继承了Debian稳定而开放的传统，很适合在进行涉及到网络安全的项目时使用。
